```python
import pandas as pd
import numpy as np
from itertools import product
```

```python
prices = ['30','10','05']
time_insulated = ['0.5', '1', '3']
capacity = ['12', '20', '32']
cleanability = ['D', 'F','E']
containment = ['Sl', 'Sp', 'Lk']
```

```python
df = pd.DataFrame(list(product(prices, time_insulated,capacity,cleanability,containment )), 
                  columns=['Pr', 'In','Cp','Cl','Cn'])
df.index += 1 
print(df.shape)
df.head()
```

```python
df_ohe = pd.get_dummies(df)
df_ohe.head()
```

```python
df_ohe.columns
```

```python
# Getting the right order
df_ohe = df_ohe[['Pr_30','Pr_10','Pr_05','In_0.5','In_1','In_3','Cp_12','Cp_20','Cp_32','Cl_D', 'Cl_F', 'Cl_E','Cn_Sl','Cn_Sp','Cn_Lk']]
```

```python
df_customers = pd.read_csv('https://raw.githubusercontent.com/Kenrich005/ISB_Marketing_Analytics/main/mugs-preference-parameters-full.csv')
df_customers.head()
```

```python
df_customers.columns
```

```python
df_customers.columns = df_customers.columns.str.strip()
df_customers.columns
```

```python
# We don't need the Brand columns
brand_cols = list(filter(lambda k: 'Br' in k, df_customers.columns))
brand_cols
```

```python
# Dropping these columns
df_customers = df_customers.drop(columns = brand_cols)
df_customers.head()
```

```python
df_customers.columns
```

```python
# Importance Matrix

df_importance = pd.DataFrame()
df_importance['IPr30'] = df_customers['pPr30']*df_customers['IPr']
df_importance['IPr10'] = df_customers['pPr10']*df_customers['IPr']
df_importance['IPr05'] = df_customers['pPr05']*df_customers['IPr']

df_importance['Iin0.5'] = df_customers['pIn0.5']*df_customers['Iin']
df_importance['Iin1'] = df_customers['pIn1']*df_customers['Iin']
df_importance['Iin3'] = df_customers['pIn3']*df_customers['Iin']

df_importance['ICp12'] = df_customers['pCp12']*df_customers['ICp']
df_importance['ICp20'] = df_customers['pCp20']*df_customers['ICp']
df_importance['ICp32'] = df_customers['pCp32']*df_customers['ICp']

df_importance['IpClD'] = df_customers['pClD']*df_customers['ICl']
df_importance['IpClF'] = df_customers['pClF']*df_customers['ICl']
df_importance['IpClE'] = df_customers['pClE']*df_customers['ICl']

df_importance['ICnSl'] = df_customers['pCnSl']*df_customers['Icn']
df_importance['ICnSp'] = df_customers['pCnSp']*df_customers['Icn']
df_importance['ICnLk'] = df_customers['pCnLk']*df_customers['Icn']
```

```python
df_importance.head()
```

```python
c = 100/(12*(644-82))
c
```

```python
prod_cust_matrix = pd.DataFrame(np.exp(np.dot(df_importance, df_ohe.T)*c))
prod_cust_matrix
```

```python
exp_sum = prod_cust_matrix.sum(axis=1)
exp_sum
```

```python
prod_cust_probs = round(prod_cust_matrix.div(exp_sum, axis=0),4)
prod_cust_probs
```

```python
prod_shares = prod_cust_probs.mean(axis=0)
prod_shares.index += 1 
prod_shares
```

```python
product_by_cust = prod_cust_probs.idxmax(axis=1)
product_by_cust
```

```python
# Most popular product
product_by_cust.value_counts()
```

```python
df_prices = df.copy()

df_prices.head()
```

```python
df_prices.info()
```

```python
df_prices['Pr'] = pd.to_numeric(df_prices['Pr'])
df_prices['In'] = pd.to_numeric(df_prices['In'])
df_prices['Cp'] = df_prices['Cp'].replace({'12':1,'20':2.6,'32':2.8})
df_prices['Cl'] = df_prices['Cl'].replace({'D':1,'F':2.2,'E':3})
df_prices['Cn'] = df_prices['Cn'].replace({'Sl':0.5,'Sp':0.8,'Lk':1})

df_prices['Profit'] = df_prices['Pr'] - df_prices['In'] - df_prices['Cp'] - df_prices['Cl'] - df_prices['Cn']
```

```python
df_prices['share'] = prod_shares
```

```python
df_prices
```

```python
df_prices['exp_profit'] = df_prices['Profit']*df_prices['share']
df_prices
```

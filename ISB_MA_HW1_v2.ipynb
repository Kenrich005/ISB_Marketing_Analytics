```python
import pandas as pd
import numpy as np
from itertools import product
```

```python
# Customer data

df_customers = pd.read_csv('https://raw.githubusercontent.com/Kenrich005/ISB_Marketing_Analytics/main/mugs-preference-parameters-full.csv')
df_customers.columns = df_customers.columns.str.strip()
df_customers.head()
```

```python
# Product data

prod1 = ['30','3','20','E', 'Lk','A']
prod2 = ['10','1','20','F', 'Sp','B']
prod3 = ['30','1','20','E', 'Lk','C']
```

```python
df_prods = pd.DataFrame(zip(prod1, prod2, prod3)).T
df_prods.columns = ['pr','in','cp','cl','cn','br']
df_prods
```

```python
df_ohe = pd.get_dummies(df_prods)
df_ohe
```

```python
df_customers.columns
```

```python
# Importance Matrix

df_importance = pd.DataFrame()

df_importance['IPr10'] = df_customers['pPr10']*df_customers['IPr']
df_importance['IPr30'] = df_customers['pPr30']*df_customers['IPr']

df_importance['Iin1'] = df_customers['pIn1']*df_customers['Iin']
df_importance['Iin3'] = df_customers['pIn3']*df_customers['Iin']

df_importance['ICp20'] = df_customers['pCp20']*df_customers['ICp']

df_importance['IpClE'] = df_customers['pClE']*df_customers['ICl']
df_importance['IpClF'] = df_customers['pClF']*df_customers['ICl']

df_importance['ICnLk'] = df_customers['pCnLk']*df_customers['Icn']
df_importance['ICnSp'] = df_customers['pCnSp']*df_customers['Icn']

df_importance['IpBrA'] = df_customers['pBrA']*df_customers['IBr']
df_importance['IpBrB'] = df_customers['pBrB']*df_customers['IBr']
df_importance['IpBrC'] = df_customers['pBrC']*df_customers['IBr']

df_importance.head()
```

```python
c = 0.0139
prod_cust_matrix = pd.DataFrame(np.exp(np.dot(df_importance, df_ohe.T)*c))
prod_cust_matrix
```

```python
exp_sum = prod_cust_matrix.sum(axis=1)
prod_cust_probs = round(prod_cust_matrix.div(exp_sum, axis=0),4)
prod_cust_probs
```

```python
prod_shares = prod_cust_probs.mean(axis=0)
prod_shares
```

```python
df_prods
```

```python
cp_cost = {'20':2.6, '12':1,'32':2.8}
cl_cost = {'D':1, 'F':2.2,'E':3}
cn_cost = {'Lk':1, 'Sp':0.8,'Sl':0.5}
```

```python
df_prices = df_prods.copy()
```

```python
df_prices = df_prices.replace({"cp": cp_cost,"cl":cl_cost, "cn":cn_cost})
df_prices['pr'] = pd.to_numeric(df_prices['pr'])
df_prices['in'] = pd.to_numeric(df_prices['in'])
df_prices
```

```python
df_prices.info()
```

```python
df_prices['cost'] = df_prices['in'] + df_prices['cp'] + df_prices['cl'] + df_prices['cn']
df_prices['margin'] = df_prices['pr'] - df_prices['cost']
df_prices['share'] = prod_shares
df_prices['exp_profit'] = df_prices['margin']*df_prices['share']
df_prices
```
